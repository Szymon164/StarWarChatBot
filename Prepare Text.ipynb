```python
import pandas as pd
import json
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
intents = json.loads(open('starwarsintents.json').read())
```

```python
words = []
classes = []
corpus = []
lemmatizer = WordNetLemmatizer()
stop_words= stopwords.words('english')

for intent in intents['intents']:
    for pattern in intent['patterns']:
        word_list = word_tokenize(pattern)
        word_list = [lemmatizer.lemmatize(word.lower()) for word in word_list if word not in words and word.isalpha() and word not in stop_words]
        corpus.append((' '.join(word_list)))
        classes.append(intent['tag'])
```

```python
vectorizer = CountVectorizer()
data = pd.DataFrame([corpus,classes],index=['data', 'label']).transpose()
```

```python
data = pd.DataFrame(vectorizer.fit_transform(data['data']).toarray())
data['label']=classes
data
```
